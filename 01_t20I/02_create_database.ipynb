```python
def create_database(data_dir=data_dir):
    years=range(2005, 2022, 1)
    #years=range(2005, 2007, 1)
    
    df_bat_all  = pd.DataFrame({}, columns=['batsman', '0s', '1s', '2s', '3s', '4s', '5s', '6s', '7s',
                                            'Runs', 'BF', 'NO', 'Team', 'Against', 'Tied', 'Win',
                                            'Toss','team-total', 'season', 'date', 'match-id' ])
    
    df_bowl_all = pd.DataFrame({}, columns=['bowler', 'O', 'M', 'Runs', 'W', 'ovs',
                                            '0s', '1s', '2s', '3s', '4s', '5s', '6s', '7s',
                                            'WD', 'NB', 'Team', 'Against', 'Tied', 'Win', 'Toss',
                                            'team-total', 'season', 'date', 'match-id'] )
    
    for year in years:
        df_match=get_match_list(year, data_dir=data_dir)
        match_ids = df_match['match_id']
        
        for match_id in match_ids:
            print (match_id)
            df_bat, df_bowl = print_scorecard(match_id+'.yaml', data_dir=data_dir)
            df_bat_all  = pd.concat([df_bat_all,  df_bat] , ignore_index=True)
            df_bowl_all = pd.concat([df_bowl_all, df_bowl], ignore_index=True)

        print (year,'Done.')

    savename_bat  =  '../database/01_t20s/batting.df'
    savename_bowl =  '../database/01_t20s/bowling.df'
        
    #savename_bat  =  '../database/01_t20s/batting_tmp.df'
    #savename_bowl =  '../database/01_t20s/bowling_tmp.df'
    
    # save the dataframes
    with open(savename_bat, 'wb') as fbat:
        pickle.dump(df_bat_all, fbat)

    with open(savename_bowl, 'wb') as fbowl:
        pickle.dump(df_bowl_all, fbowl)

#create_database()
```
