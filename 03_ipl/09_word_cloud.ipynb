```python
from wordcloud import WordCloud, STOPWORDS

def PlotWordCloud(df_bat, df_bowl=None, col='all'):
    
    if col=='all':
        names=list(df_bat['batsman'].values) + list(df_bowl['bowler'].values)
    elif col=='batsman':
        names=list(df_bat['batsman'].values)
    elif col=='bowler':
        names=list(df_bowl['bowler'].values)
    
    
    stopwords= set(STOPWORDS)
    wordcloud = WordCloud(background_color='white', max_words=1000,
                          stopwords=stopwords).generate(" ".join([b for b in names]))

    plt.figure(figsize=(20,6), facecolor='maroon', edgecolor='blue')
    plt.imshow(wordcloud)#, cmap='viridis')
    plt.axis("off")
    #plt.title("IPL Batsman", color='maroon', fontsize=34)

df_bat  = pickle.load(open('./database/batting_record.df', 'rb'))
df_bowl = pickle.load(open('./database/bowling_record.df', 'rb'))

PlotWordCloud(df_bat, df_bowl, col='batsman')
```
