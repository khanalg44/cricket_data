```python
def bating_details(player):
    df=pickle.load(open('./database/batting_record_all_years.df', 'rb'))
    dfp=df[df['batsman']==player]
    
    teams_played = dfp.Team.unique()
    
    data = []
    for team in teams_played:
        dft = dfp[dfp['Team']==team]

        Inns_ = dft.shape[0]
        Runs_ = dft.Runs.sum()
        BF_   = dft.BF.sum()
        NOs_  = dft.NO.sum()
        
        HS    = max(dft.Runs)
        Fifty = ((dft.Runs>=50) & (dft.Runs<100) ).sum()
        Hundred = (dft.Runs>=100).sum()

        Wins  = sum(dft.Win)
        Toss_wins = sum(dft.Toss)

        SR    = np.round(100*Runs_/(BF_+0.1), 2) # add 0.1 to avoid deviding by 0
        Ave   = np.round(Runs_/(Inns_-NOs_), 2)
        Fours = dft['4s'].sum()
        Sixes = dft['6s'].sum()
        
        seasons = dft['season'].unique()
        
        data.append([team, Inns_, NOs_, Runs_, BF_, HS, Ave, SR, Fifty, Hundred, Fours, Sixes, seasons] )
    df_p = pd.DataFrame(data, columns=['Team', 'Innings', 'NO', 'Runs', 'BF', 'HS',
                                       'Ave','SR', '50s', '100s', '4s', '6s', 'Seasons'])
    return df_p
```
