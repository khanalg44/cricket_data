```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, glob, yaml, pickle
import datetime

from utils import get_match_list
from print_scorecard import print_scorecard

data_dir='../datasets/ipl/yaml/'
```

```python
def create_database():
    years=range(2008, 2021, 1)
    #years = [2008]
    df_bat_all  = pd.DataFrame({}, columns=['batsman', '0s', '1s', '2s', '3s', '4s', '5s', '6s',
                                            'Runs', 'BF', 'NO', 'Team', 'Against', 'Win',
                                            'Toss','team-total', 'season', 'date', 'match-id' ])
    
    df_bowl_all = pd.DataFrame({}, columns=['bowler', 'O', 'M', 'R', 'W',
                                            '0s', '1s', '2s', '3s', '4s', '5s', '6s',
                                            'WD', 'NB', 'Team', 'Against', 'Win', 'Toss',
                                            'team-total', 'season', 'date', 'match-id'] )
    for year in years:
        df_match=get_match_list(year)
        match_ids = df_match['match_id']
        
        for match_id in match_ids:
            df_bat, df_bowl =print_scorecard(match_id+'.yaml', data_dir=data_dir)
            df_bat_all  = pd.concat([df_bat_all,  df_bat] , ignore_index=True)
            df_bowl_all = pd.concat([df_bowl_all, df_bowl], ignore_index=True)

        print (year,'Done.')
        
    # save the dataframe
    with open('./database/batting_record.df', 'wb') as fo:
        pickle.dump(df_bat_all, fo)

    with open('./database/bowling_record.df', 'wb') as fo:
        pickle.dump(df_bowl_all, fo)

        
#create_database()
```
